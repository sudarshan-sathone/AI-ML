## Python Fundamentals

In this section, I implemented basic Python concepts like variables, loops, and functions.
The function acts as a mini-processor that analyzes numerical data and returns useful statistics like minimum, maximum, and average.

This demonstrates how AI systems process and validate raw data before analysis.

```python
name = "Developer_Sudarshan"
age = 24
marks = [85, 92, 78, 95, 88]

for i in marks:
    print(f"Validating Data Point: {i}")

def analyze_numbers(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    avg_val = sum(numbers) / len(numbers)
    return min_val, max_val, avg_val

results = analyze_numbers(marks)

print("\n--- Statistics Report ---")
print(f"Low: {results[0]} | High: {results[1]} | Average: {results[2]}")
```

## NumPy Operations

NumPy is used for handling numerical data efficiently. I used NumPy arrays instead of Python lists because they allow vectorized operations.

Vectorization means performing operations on entire datasets at once without using loops. This makes computations faster and is very important in AI models.

I reshaped a 1D array into a 2x2 matrix, which is similar to how image data is structured in machine learning.

```python
import numpy as np

data_points = np.array([10, 20, 30, 40])

matrix = data_points.reshape(2, 2)
print("Matrix:\n", matrix)

processed_data = data_points * 2
print("Scaled Data:", processed_data)
```
