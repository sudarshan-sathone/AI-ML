```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_breast_cancer
import joblib

data = load_breast_cancer()
X, y = data.data, data.target
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

rf = RandomForestClassifier(random_state=42)

grid = GridSearchCV(rf, param_grid, cv=3, scoring='accuracy')
grid.fit(X, y)

print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
joblib.dump(grid.best_estimator_, "best_model.pkl")
```

### Reflection

GridSearchCV tests every possible combination of hyperparameters.

If we have 5 parameters with 10 values each:
Total combinations = 10^5 = 100,000 models

With cross-validation, this number increases further.

This becomes computationally expensive for large datasets.

RandomizedSearchCV is preferred in such cases because it selects a random subset of parameter combinations, reducing training time while still finding good results.
