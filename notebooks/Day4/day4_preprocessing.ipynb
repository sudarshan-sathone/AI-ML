# Day 4: Pre-processing Techniques

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [21, None, 23, 24],
    'Score': [80, 75, None, 95]
}

df = pd.DataFrame(data)
df
```

```python
df_dropped = df.dropna()
df_dropped
```

```python
df_filled = df.copy()

df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].mean())
df_filled['Score'] = df_filled['Score'].fillna(0)

df_filled
```

## Reflection

Filling missing values is better than dropping rows because dropping removes valuable data and reduces dataset size.

Using mean for numerical values like age maintains realistic distribution, whereas filling with 0 introduces incorrect data and biases the model.

Thus, imputation helps preserve both data quantity and quality.
