# Day 11: Random Forests & Feature Importance

In this task, I implemented a Random Forest Regressor and analyzed how the number of trees affects model performance and training time.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import time

# Load dataset
data = fetch_california_housing()

X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)
```

```python
tree_counts = [10, 50, 200]

results = []

for n in tree_counts:
    start = time.time()

    model = RandomForestRegressor(n_estimators=n, random_state=42)
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)
    score = r2_score(y_test, predictions)

    end = time.time()

    results.append((n, score, end - start))

    print(f"Trees: {n}")
    print(f"R2 Score: {score:.4f}")
    print(f"Training Time: {end - start:.2f} seconds\n")
```

## Reflection

As I increased the number of trees in the Random Forest, I observed that the model’s performance improved initially, especially when moving from 10 to 50 trees. However, beyond a certain point, the improvement in R² score became very minimal while the training time increased significantly.

This demonstrates the concept of diminishing returns, where increasing complexity does not always lead to proportional gains in performance. Therefore, it is important to choose an optimal number of trees that balances both accuracy and efficiency.
