```python
from textblob import TextBlob

bios = [
    "I love attending tech events, amazing experience!",
    "This event was terrible and badly organized.",
    "It was okay, not great but not bad.",
    "Absolutely fantastic networking opportunity!",
    "Worst experience ever, waste of time."
]

for bio in bios:
    blob = TextBlob(bio)
    print("Text:", bio)
    print("Polarity:", blob.sentiment.polarity)
    print("Subjectivity:", blob.sentiment.subjectivity)
    print("-" * 50)
```

## Reflection

MeetMux can use sentiment analysis on event reviews to automatically detect negative feedback patterns. If multiple users consistently give negative polarity scores for a specific organizer, the system can flag them for poor user experience. This helps maintain platform quality and enables quick action without manual review.

## Stemming vs Lemmatization

- Stemming: Cuts words (e.g., running → runn)
- Lemmatization: Converts to meaningful root word (e.g., running → run)
