Day 15: Unsupervised Learning & User Segmentation

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=300, centers=5, cluster_std=0.60, random_state=0)

print("Run 1 (k-means++):")
k1 = KMeans(n_clusters=5, init='k-means++', random_state=42).fit(X)
print(k1.cluster_centers_)

print("\nRun 2 (random):")
k2 = KMeans(n_clusters=5, init='random', random_state=42).fit(X)
print(k2.cluster_centers_)

print("\nRun 3 (random different seed):")
k3 = KMeans(n_clusters=5, init='random', random_state=10).fit(X)
print(k3.cluster_centers_)
```

In this experiment, the cluster centers across different runs appear similar, though their order changes. This indicates that the dataset is well-structured and K-Means converges to the same solution even with random initialization.

However, in general, random initialization can lead to different results depending on the dataset. The k-means++ method improves stability by selecting better initial centroids, reducing the chances of poor clustering and ensuring more consistent results.
